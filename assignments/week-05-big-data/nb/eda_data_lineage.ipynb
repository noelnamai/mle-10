```python
for company in ["AAPL", "MSFT", "AMZN", "TSLA", "GOOGL"]:
    url = f"https://www.alphavantage.co/query?function=EARNINGS&symbol={company}&apikey=secret_keys.apikey"
    results = requests.get(url)
    company_data = results.json()
    time.sleep(15)

    annual_earnings_df = pd.DataFrame(company_data["annualEarnings"])
    annual_earnings_df.columns = ["date", company.lower() + "_annual_eps"]
    annual_earnings_df["date"] = pd.to_datetime(annual_earnings_df["date"])

    quarterly_earnings_df = pd.DataFrame(company_data["quarterlyEarnings"])
    quarterly_earnings_df = quarterly_earnings_df[["reportedDate", "reportedEPS", "surprise"]]
    quarterly_earnings_df.columns = ["date", company.lower() + "_quarterly_eps", company.lower() + "_surprise"]
    quarterly_earnings_df["date"] = pd.to_datetime(quarterly_earnings_df["date"])

    earnings_df = pd.merge(annual_earnings_df, quarterly_earnings_df, how="outer")
    df = pd.merge(df, earnings_df, how="outer")
```

```python
df = df[df["date"] >= "2002-01-01"]

df.sort_values(by="date", inplace=True, ignore_index=True)

df.replace(".", np.nan, inplace=True)
df.replace("None", np.nan, inplace=True)

df.fillna(method="ffill", inplace=True)
df.fillna(method="bfill", inplace=True)

cols = df.columns.drop(["date"])
df[cols] = df[cols].apply(pd.to_numeric)

df["gap"] = df["open"] - df["close"].shift()
```

```python
df["day"] = df["date"].dt.day_name()
df["month"] = df["date"].dt.month_name()
df["year"] = df["date"].dt.year
```
