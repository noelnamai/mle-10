## Task 3: Preprocess Data

Correlated features in general don't improve models (although it depends on the specifics of the problem like the number of variables and the degree of correlation), but they affect specific models in different ways and to varying extents:

- For linear models (e.g., linear regression or logistic regression), [multicolinearity](https://en.wikipedia.org/wiki/Multicollinearity) can yield [solutions that are wildly varying and possibly numerically unstable](https://en.wikipedia.org/wiki/Multicollinearity#Consequences).
- Random forest can be good at detecting interactions between different features, but highly correlated features can mask these interactions.
More generally, this can be viewed as a special case of [Occam's razor](https://en.wikipedia.org/wiki/Occam%27s_razor). A simpler model is preferable, and, in some sense, a model with fewer features is simpler. The concept of [minimum description length](https://en.wikipedia.org/wiki/Minimum_description_length) makes this more precise ([ref](https://datascience.stackexchange.com/questions/24452/in-supervised-learning-why-is-it-bad-to-have-correlated-features)).

2. Drop the features with high correlation.

    In the "Correlations" tab in the profile report, you can find heatmap using five different correlation metrics; toggle correlation desciptions if any one of them looks foreign to you.  

    We'll remove high correlated features by looking at each pair of features, and if they are highly correlated (at least 0.8), we won't include the second feature in the pair. Store the remaining set of features (the ones you didn't drop) in dataframes `train_df_reduced`. You are given most the code, make sure that you understand what each line does.
    
    First, calculate the correlation with just 10,000 samples.

```python
# YOUR CODE HERE
train_small_df = train_df.copy().sample(n=int(1e4), random_state=12) 
cor = train_small_df.corr(numeric_only=True)
```

```python
import numpy as np

keep_columns = np.full(cor.shape[0], True)

for i in range(cor.shape[0] - 1):
    for j in range(i + 1, cor.shape[0] - 1):
        if (np.abs(cor.iloc[i, j]) >= 0.8):
            keep_columns[j] = False

selected_columns = train_df.columns[keep_columns]
train_df_reduced = train_df[selected_columns]
```

### Algorithm Understanding.
    
    1. Is SVM (Support Vector Machine) a supervised or unsupervised learning algorithm?
    
    - Support Vector Machine (SVM) is a supervised learning algorithm since it requires labeled training data.

    2. Why is SVM such a powerful classification method?

    - SVM is effective in high-dimensional spaces and with a large number of training examples since it can maximize the distance between the decision boundary and the closest points of each class.
    - Use the kernel trick to find nonlinear decision boundaries espeacialy when the data is not linearly separable.
    - SVM has a regularization parameter which helps prevent overfitting and improve the generalization ability of the model.

    3. What are 3 disadvantages of SVMs?

    - SVMs are very sensitive to the choice of kernel and the kernel's parameters. This could be computational expensive for large and complex datasets.
    - The training time for SVM can be long and hence difficult to scale SVM to very large datasets. This makes SVM less suitable for real-time applications where fast performance is required.
    - SVM may produce suboptimal results when dealing with imbalanced datasets and may not be able to find a good decision boundary that separates the two classes.

    3. Explain feature importance for the Random Forest algorithm?

    - Feature importance is the relative importance of each feature in predicting the target variable.
    - The algorithm computes feature importance by considering how much each feature contributes to the overall decision-making process of the model and assigns a score to each feature accordingly.
    - Features with higher scores are considered to be more important than those with lower scores.

    4. When examining feature importance, what is Gini impurity or information gain?

    - It is the probability of a random sample being misclassified and is calculated as the sum of the probability of each class in the subset multiplied by the probability of not being in that class.
    - Information gain is a measure of the decrease in entropy that results from splitting the data on a particular feature. It is calculated as the difference between the entropy of the parent node and the weighted sum of the entropies of the child nodes. The higher the information gain, the more important the feature is considered to be.

    5. SHAP (SHapley Additive exPlanations) is a game theoretic approach to explain the output of any machine learning model, what is it and how does it work?

    - SHAP is a method for explaining the output of a machine learning mode and assigns a measure of importance to each feature indicating how much that feature contributes to the model's output. These measures of importance are determined using a game-theoretic approach known as the Shapley value.
